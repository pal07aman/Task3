```python
X_train=np.zeros((len(train_files),128,128))
Y_train=np.zeros((len(train_files),1))
for i, fi in enumerate(train_files):
    im=Image.open(fi)
    # : -> all elements
    X_train[i,:,:]=np.array(im)
    Y_train[i]=  fi[-6]
```

```python
Y_train=Y_train-1
```

```python
X_train
```

```python
X_train/255.0
```

```python
import keras
```

```python
from tensorflow.keras.layers import Conv2D
```

```python
from keras.models import Sequential
from keras.layers import Dense , Flatten , Conv2D , LSTM, Dropout 
from keras.layers import BatchNormalization
```

```python
from keras.layers import MaxPool2D
```

```python
model=keras.Sequential()

model.add(Conv2D(32,(3,3),strides=2,padding='Same',activation='relu',input_shape
=(128, 128, 1)))
model.add(MaxPool2D(2,2))
model.add(BatchNormalization())
model.add(Dropout(0.3)) #here added dropout
model.add(Conv2D(64,(3,3), strides=1,padding= 'Same', activation='relu'))
model.add(MaxPool2D(2,2))
model.add(BatchNormalization())
model.add(Conv2D(128,(3,3), strides=1,padding= 'Same', activation='relu'))
model.add(BatchNormalization())
model.add(Conv2D(256,(3,3), strides=1,padding= 'Same', activation='relu'))
model.add(MaxPool2D(2,2))
model.add(BatchNormalization())
model.add(Flatten())
model.add(Dropout(0.5))
model.add(Dense(512, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(3, activation = "softmax"))
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.summary()
```

```python
X_train=X_train.reshape(len(X_train),128,128,1)
```

```python
model.fit(X_train, Y_train,epochs=2,validation_split=0.1)
```

```python
prediction = model.predict(
    X_train,
    batch_size= None,
    verbose= 0,
    steps = None,
    callbacks = None,
    max_queue_size = 10,
    workers = 1,
    use_multiprocessing = False
)
```
